```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv('LSTM-Multivariate_pollution.csv')
df.dropna(inplace=True)
```

```python
non_numeric_columns = df.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_columns)
```

```python
for col in non_numeric_columns:
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
df.dropna(inplace=True)
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df.iloc[:, 1:].values)
```

```python
def create_sequences(data, seq_length):
    sequences, labels = [], []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i + seq_length])
        labels.append(data[i + seq_length, 0])  # Predicting first feature
    return np.array(sequences), np.array(labels)

seq_length = 10  # Adjust based on requirement
X, y = create_sequences(data_scaled, seq_length)
```

```python
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
model = Sequential([
    Bidirectional(LSTM(50, return_sequences=True), input_shape=(seq_length, X.shape[2])),
    Bidirectional(LSTM(50)),
    Dense(25, activation='relu'),
    Dense(1)
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# Train model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Plot training loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()

# Evaluate model
test_predictions = model.predict(X_test)
plt.plot(y_test, label='Actual')
plt.plot(test_predictions, label='Predicted')
plt.legend()
plt.show()
```
